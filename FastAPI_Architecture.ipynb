# Architecture, Design Patterns & Request-Response Life Cycle

This section is extra reading for every FastAPI developer. Simple explanations and examples are included.

## How FastAPI is Built (Architecture)

- ASGI: FastAPI uses ASGI, a modern async interface between the server and your app.

   - Think of it as a pipe: the server sends a request → your app sends a response back.

- Under the hood:

   - Starlette: Handles web requests, responses, and routing.

   - Pydantic: Validates and shapes data automatically.

- Main components:

   - App: One per project, main entry point.

   - Routers: Groups of routes, usually by feature.

   - Dependencies: Reusable pieces, e.g., “get DB session” or “current user”.

   - Middleware: Code that runs for every request before and after your routes.

- Request flow:

> Request comes in → Middleware runs → Route selected → Dependencies run → Route code executes → Response passes back through middleware

## How to Structure Your Code (Design Patterns)

- Layered approach:

   - Routes: Keep them simple. Just accept request → call service function → return response.

   - Business logic & database access: Keep in separate modules, not inside routes.

- Dependency Injection:

   - Use Depends() to inject shared resources like DB sessions or current user.

   - FastAPI executes these dependencies before the route automatically.

- Routers:
 
    - Group related routes by feature (products, categories, auth).

    - Attach each group to the app with a prefix, e.g., /products.

- Separation of Concerns:

    - Pydantic schemas → define request and response shapes.

    - SQLAlchemy models → define database tables.

    - Routes connect the two and return responses.

## Request-Response Life Cycle
**When a request arrives:**

**1. Middleware runs first**

- Example: log the request, add request ID, measure execution time.

**2. Routing**

- FastAPI matches URL + HTTP method to a route.

**3. Dependencies run**

- Example: “get DB session” → “get current user”.

- If a dependency fails, the route does not run.

**4. Validation occurs**

- Body, query, and path parameters are validated automatically.

- Invalid data → FastAPI returns 422, route is skipped.

**5.Route code runs**

- Route executes and returns a value.

**When the response goes out:**

1. Route returns a value (dict, list, or Pydantic model).

2. FastAPI converts it to JSON (using response_model if defined).

3. Response passes back through middleware → can log or modify response.

**Important Execution Order**
> Middleware → Routing → Dependencies & Validation → Route → Middleware (response)

- Middleware sees every request and response.

- Dependencies execute before route.

- Validation errors prevent the route from running.
