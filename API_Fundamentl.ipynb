# API Fundamentals

## 1. What is an API and Why We Use It

**Definition:**  
An **API (Application Programming Interface)** is a set of rules and protocols that lets one software program communicate with another.  

**Analogy:**  
- You (client) send a request: “I want spaghetti”  
- The waiter (API) delivers it to the kitchen (server)  
- The kitchen prepares it and the waiter brings it back to you  

**Why APIs are used:**

| Purpose       | Description                                                                 | Example                                               |
|---------------|-----------------------------------------------------------------------------|------------------------------------------------------|
| Integration   | Different systems or apps can work together                                 | Weather app uses API to get data from a weather service |
| Data Access   | Controlled access to data without direct database access                    | Accessing user info without exposing DB             |
| Automation    | Enables machines/software to interact automatically                        | Automatically posting social media updates         |
| Consistency   | Standardizes communication, making development easier and scalable         | Using the same endpoints across multiple apps      |

## 2. REST APIs

**REST (Representational State Transfer)** is the most common API style. It treats everything as a **resource**.

### 1. Resources
- Anything you can interact with is a **resource**.  
- Examples: Users, Products, Orders, Messages  
- Each resource has a URL/endpoint:
https://api.example.com/users/123


Here, `users` is the resource, `123` is the resource ID.

### 2.  HTTP Methods (Actions on Resources)

| Method  | Action           | Example                           |
|---------|-----------------|----------------------------------|
| GET     | Retrieve data    | GET /users/123                   |
| POST    | Create resource  | POST /users                      |
| PUT     | Replace resource | PUT /users/123                   |
| PATCH   | Partial update   | PATCH /users/123                 |
| DELETE  | Remove resource  | DELETE /users/123                |

**Key difference:**  
- **PUT** replaces the entire resource.  
- **PATCH** updates only specified fields.

## 3. Requests and Responses

### 1.  Request Components

| Component       | Description                                | Example |
|-----------------|--------------------------------------------|---------|
| URL / Endpoint  | Identifies the resource                     | https://api.example.com/products/45 |
| HTTP Method     | Action to perform (GET, POST, etc.)        | GET /products/45 |
| Headers         | Metadata about the request                  | Content-Type: application/json <br> Authorization: Bearer `<token>` |
| Body (optional) | Data sent to server (for POST/PUT/PATCH)   | ```json { "name": "Bismah", "email": "bismah@example.com" } ``` |

### 2.  Response Components

| Component    | Description                                 | Example |
|-------------|--------------------------------------------|---------|
| Status Code | Tells result of request                     | 200 OK, 201 Created, 404 Not Found |
| Headers     | Metadata about response                     | Content-Type: application/json    |
| Body        | Main content, usually in JSON               | ```json { "id": 123, "name": "Bismah", "email": "bismah@example.com" } ``` |

**Common HTTP Status Codes**

| Code | Meaning       |
|------|---------------|
| 200  | OK / Success  |
| 201  | Created       |
| 400  | Bad Request   |
| 401  | Unauthorized  |
| 403  | Forbidden     |
| 404  | Not Found     |
| 500  | Server Error  |

---

## 4. Authentication vs Authorization

| Concept       | Purpose                       | Example |
|---------------|-------------------------------|---------|
| Authentication | Verify identity (who you are) | Logging in with email & password, or using API key/token |
| Authorization  | Check permissions (what you can do) | Admin can delete users, normal user cannot |

**Analogy:**  
- Authentication = “Are you really Bismah?”  
- Authorization = “Are you allowed to delete this post?”
