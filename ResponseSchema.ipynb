## Defining Request Body Schemas with Pydantic
Define Request body schema using BaseModel pydantic class

### Example

```python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()
```

```python
# Request / Response Model
class User(BaseModel):
    name: str
    age: int
```

```python
@app.post("/users/")
def create_user(user: User): # here define request schman
    return {"message": f"{user.name} is {user.age} years old"}
```

### Defining Response Schemas with response_model

Pydantic models can also define the structure of the API response. This ensures that clients always get data in the expected format, even if your internal data contains extra fields.

### Example

```python
@app.get("/users/{user_id}", response_model=User) # here define response schema
def get_user(user_id: int):
    db_user = {"name": "Alice", "age": 30, "password": "secret"}
    return db_user
```

### Basic Field Types, Defaults, and Optional Fields

Pydantic allows you to define:

- **Field types:** str, int, float, bool, list, dict, etc.

- **Defaults:** Use = to provide default values.

- **Optional fields:** Use Optional[type] for fields that may or may not be provided.

## How FastAPI Uses Pydantic for Validation and Docs

### Validation

FastAPI uses Pydantic models to:

1. Check types: e.g., integers, strings, floats, booleans.

2. Enforce required fields: fields without defaults must be provided.

3. Apply constraints: You can add things like min_length, max_length, ge, le for numbers.

4. Return structured error messages automatically.

```python
# Example of validation error:

{
  "detail": [
    {
      "loc": ["body", "age"],
      "msg": "value is not a valid integer",
      "type": "type_error.integer"
    }
  ]
}
```

### **Documentation (Swagger / OpenAPI)**

1. FastAPI reads Pydantic models to generate interactive API docs.

2. Field names, types, optionality, and defaults all appear automatically.

3. You don’t need to manually document each request or response.
