## Defining Routes and Path Parameters
### Routes
In FastAPI, a route is a URL path that your application responds to. You define it using decorators like @app.get(), @app.post(), etc.

- /hello is the route.

- GET is the HTTP method

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/hello")
def say_hello():
    return {"message": "Hello, FastAPI!"}
```

### Path Parameters
You can capture dynamic values from the URL:

```python
@app.get("/users/{user_id}")
def get_user(user_id: int):
    return {"user_id": user_id}
```

- {user_id} is a path parameter.

- FastAPI automatically converts the parameter to int because of the type hint.

- You can use multiple path parameters:

```python
@app.get("/users/{user_id}/items/{item_id}")
def get_item(user_id: int, item_id: str):
    return {"user": user_id, "item": item_id}
```

## Query Parameters and Optional Parameters
### Query Parameters

Parameters that come after ? in the URL are query parameters. FastAPI reads them automatically from the function signature.

```python
@app.get("/search/")
def search(q: str):
    return {"query": q}
```

### Optional Query Parameters

You can make parameters optional with Optional or default values:

```python
from typing import Optional

@app.get("/search/")
def search(q: Optional[str] = None, limit: int = 10):
    return {"query": q, "limit": limit}
```

### Request Body and FastAPI Binding

When clients send JSON, form data, or other body content, FastAPI automatically parses it and binds it to your function parameters using Pydantic models or Python types.

- FastAPI converts JSON → Pydantic object → Python object.

- Automatic validation occurs (e.g., price must be float).

### Organizing Routes

For larger apps, it’s best to group related routes by resource or module.
#### Benefits:

- Routes are organized by resource (/users, /items, etc.).

- Easier to maintain and scale.

- Tags are used for Swagger documentation grouping.
