## Why Caching Matters for API Performance
Caching is storing a copy of the result of a computation or query so that future requests can be served faster without recomputing or refetching.


#### Benefit
- Reduced latency
- Lower server load
- Scalability

### Caching Use Cases
- Repeated queries

- External API calls

- Expensive computations


### Types of Caching and Key Considerations

| Type     | Description  |
|----------|----------|
| In-Memory Cache  | short term, store in server RAM , loss when server start | 
| Distributed Cache  | store in external server, Expiration Time  | 
| Client-Side Cache   | store on client browser , need proper cache header | 
| Database Query Cache   | Cache DB query results , invalidation on DB updates  | 


**Expiration:** Avoid stale data; set TTL (time-to-live) for cache entries.

**Invalidation:** Clear cache when data changes.

**Cache size:** Avoid memory overload; use eviction policies (LRU, LFU).

**Security:** Don’t cache sensitive user data without protection.

## Using Redis with FastAPI

```python
import redis

r = redis.Redis(host='localhost', port=6379, db=0)
```

```python
from fastapi import FastAPI
import json

app = FastAPI()

@app.get("/weather/{city}")
def get_weather(city: str):
    cached = r.get(city)
    if cached:
        return json.loads(cached)  # Return cached response

    # Simulate external API call
    data = {"city": city, "temp": 25}

    # Store in Redis for 10 minutes (600 seconds)
    r.setex(city, 600, json.dumps(data))
    return data
```

## Caching DB Queries and External API Responses in FastAPI

```python
@app.get("/users")
def get_users():
    cached = r.get("users_list")
    if cached:
        return json.loads(cached)

    # Simulate DB query
    users = [{"id": 1, "name": "Alice"}, {"id": 2, "name": "Bob"}]

    # Cache result for 5 minutes
    r.setex("users_list", 300, json.dumps(users))
    return users
```

### Summary

| Aspect            | Notes                                                                 |
|-------------------|-----------------------------------------------------------------------|
| Why cache         | Faster response, lower load, scalable API                             |
| When to cache     | Repeated queries, external API calls, expensive computations          |
| Types             | In-memory, distributed (Redis/Memcached), client-side, DB query cache|
| Key points        | TTL, invalidation, cache size, security                               |
| FastAPI + Redis   | Connect with redis-py, use setex for TTL, fetch cached data first     |
