## Why APIs Need Authentication

APIs are endpoints that allow external systems or clients to interact with your application. 

#### Without authentication:

- Anyone can access
- Perform sensitive operations


## Common Approaches to API Authentication
### API Keys

An API key is a unique string (like a1b2c3d4e5) that identifies a client

#### How it works:

- Server generates a key for a user or app.
- The client includes this key in requests (usually in headers or as a query parameter).
- Server validates the key and allows access if valid

Mostly used when server to server communication


## JWT (JSON Web Token)
JWT is a token that contains encoded data (claims) about the user and is signed by the server.

#### **Structure:**

> header.payload.signature


**Header:** Algorithm used for signing (e.g., HS256)

**Payload:** User info (e.g., user ID, role)

**Signature:** Ensures token hasn’t been tampered with

## How it works:
1. User logs in with username/password.

2. Server generates JWT containing user info and sends it to the client.

3. Client includes JWT in requests (usually in the Authorization header).

4. Server verifies JWT signature and extracts user info to authorize access.

## Protecting Routes with Dependencies
In frameworks like FastAPI, you can attach dependencies that check authentication before allowing access to certain routes.

```python
from fastapi import Depends, HTTPException

def verify_token(token: str):
    if token != "valid_token":
        raise HTTPException(status_code=401, detail="Unauthorized")
    return True

@app.get("/protected")
def protected_route(auth: bool = Depends(verify_token)):
    return {"message": "You are authorized!"}
```

### Not Storing Secrets in Code

Why: Hardcoding API keys, passwords, or tokens in code is dangerous. If your code is pushed to GitHub or shared, secrets are exposed.

Use environment variables (.env) or secure config systems.
