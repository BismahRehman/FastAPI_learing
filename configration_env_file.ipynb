## Using Environment Variables
### Why


- Avoids hardcoding sensitive info in your code (DB passwords, API keys, secret tokens).

- Makes your app configurable per environment (dev, staging, prod).

### .env file

#### Access in Python:

```python
# Access in Python

import os

db_url = os.getenv("DATABASE_URL")
secret_key = os.getenv("SECRET_KEY")
```

### Loading Config in One Place

### Why:

- Centralizes all configuration in a single module.

- Makes your code cleaner and easier to maintain.

- Example with Pydantic Settings (FastAPI-friendly):

### in config file


```python
from pydantic import BaseSettings

class Settings(BaseSettings):
    database_url: str
    secret_key: str
    api_key: str

    class Config:
        env_file = ".env"

settings = Settings()
```

```python
from config import settings

print(settings.database_url)
```

### Different Config for Dev vs Production (Conceptual)

#### Why:

Development and production often need different DBs, API keys, debug flags.

#### Approach:

Use environment variables or separate .env files:

```python
# .env.dev
DEBUG=True
DATABASE_URL=sqlite:///./dev.db

# .env.prod
DEBUG=False
DATABASE_URL=postgresql://user:pass@prod-db/db
```

Pydantic can automatically load based on an environment variable:


```python
import os
from pydantic import BaseSettings

class Settings(BaseSettings):
    database_url: str
    secret_key: str
    api_key: str
    debug: bool = False

    class Config:
        env_file = ".env.dev" if os.getenv("ENV") == "dev" else ".env.prod"

settings = Settings()
```

### Keeping Secrets Out of Version Control

#### Problem:

If you push a .env with passwords or API keys to GitHub, anyone can access them.

#### Best Practices:

Add .env to .gitignore

## Summary – How It All Works Together

**.env**→ stores secrets and configs (DB URL, API keys, etc.)

**Settings module** → loads env variables, validates types, provides a single place for all configs

**App code** → reads settings from Settings module instead of hardcoding

**Different .env per environment** → allows dev vs prod configs

**.gitignore or secret managers** → keep secrets safe
